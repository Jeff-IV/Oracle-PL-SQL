### Bloque anonimo o bloque PL

Un bloque en Oracle esta conformado por una zona de declaracion de variables, un inicio y un final.
Cada linea agregada en el codigo debe terminar con un punto y coma.


```
DECLARE
    v_num NUMBER(2) := 10;
    v_string VARCHAR(10) := 'Jeff';
    v_date DATE :=  SYSDATE;
BEGIN
    DBMS_OUTPUT.PUT_LINE('Hello world!');
    DBMS_OUTPUT.PUT_LINE('v_num value: ' || v_num);
    DBMS_OUTPUT.PUT_LINE('v_string value: ' || v_string);
    DBMS_OUTPUT.PUT_LINE('v_date value: ' || v_date);
END;
```

## Pedir datos por consola

Para pedir o ingresar datos por consola, al crear la variable le asignaremos un & y un nombre (placeholder)

```
DECLARE
    v1 NUMBER(5) := &valor_1;
    v2 NUMBER(5) := &valor_2;
    v_sum NUMBER (10);
BEGIN
    v_sum := v1 + v2;
    DBMS_OUTPUT.PUT_LINE(v1 || ' + ' || v2 || ' = ' || v_sum);
END;
```

## Bloques if elsif y else


Podemos manejar el flujo con el bloque if, veremos como se implementa y los diferentes operadores logicos

```
DECLARE 
    v1 NUMBER(2) := 1;
    v2 NUMBER(2) := 1;
    v3 NUMBER(2) := 1;
BEGIN
    IF v1 > v2 AND v1 > v3 THEN
    DBMS_OUTPUT.PUT_LINE(v1 || ' is the largest');
    ELSIF v2 > v1 AND v2 > v3 THEN
    DBMS_OUTPUT.PUT_LINE(v2 || ' is the largest');
    ELSIF v1 = v2 AND v1 = v3 THEN
    DBMS_OUTPUT.PUT_LINE(v1 || ' and ' || v2 || ' and ' || v3 || ' are equals');
    ELSE
    DBMS_OUTPUT.PUT_LINE(v3 || ' is the largest');
    END IF;
END;
```

## Operadores logicos


En Oracle los operadores logicos son casi los mismos que en python con la distincion de que el distinto de o diferente de es <> y el de igualdas es =


```
DECLARE 
    v1 NUMBER(2) := 3;
    v2 NUMBER(2) := 1;
BEGIN
    IF v1 <> v2  THEN
    DBMS_OUTPUT.PUT_LINE(v1 || ' and ' || v2 || ' are differents');
    ELSIF v1 = v2 THEN
    DBMS_OUTPUT.PUT_LINE(v1 || ' and ' || v2 || ' are equals');
    END IF;
END;
```

## Bloque Case


El bloque case nos sirve para cuando tenemos que anidar varios condicionales, esto le da mas legibilidad al codigo


```
DECLARE
    v_day NUMBER(1) := &weekday;
BEGIN
    CASE v_day

    WHEN 1 THEN 
        DBMS_OUTPUT.PUT_LINE('Monday');

    WHEN 2 THEN 
        DBMS_OUTPUT.PUT_LINE('Tuesday');

    WHEN 3 THEN 
        DBMS_OUTPUT.PUT_LINE('Wednesday');

    WHEN 4 THEN 
        DBMS_OUTPUT.PUT_LINE('Thursday');

    WHEN 5 THEN 
        DBMS_OUTPUT.PUT_LINE('Friday');

    WHEN 6 THEN 
        DBMS_OUTPUT.PUT_LINE('Saturday');

    WHEN 7 THEN 
        DBMS_OUTPUT.PUT_LINE('Sunday');

    ELSE 
        DBMS_OUTPUT.PUT_LINE('Error: number out of range');
        
    END CASE;
END;
```

## Bucles


### While


```
DECLARE
    v_i NUMBER(2) := 1;
BEGIN

    WHILE (v_i <= 10)
    LOOP
        DBMS_OUTPUT.PUT_LINE(v_i);
        v_i := v_i + 1;

    END LOOP;


END;
```

### FOR


En el bucle for podemos usar una variable sin haberla declarado en el bloque de DECLARE (esta variable es local, solo existe en su respectivo loop), la variable se incrementa de manera automatica (1 en 1 por defecto)

```
DECLARE

BEGIN

    FOR v_i IN 1..10
    LOOP
        DBMS_OUTPUT.PUT_LINE(v_i);

    END LOOP;

    FOR v_initial IN REVERSE 1..10
    LOOP
        DBMS_OUTPUT.PUT_LINE(v_initial);

    END LOOP;

END;
```

### LOOP


Este bucle lo que nos permite es poner una condicion en cualquier lugar del loop

```
DECLARE
    v_i NUMBER(2) := 5;
BEGIN

    LOOP
        DBMS_OUTPUT.PUT_LINE(v_i);
        EXIT WHEN v_i = 9;
        v_i := v_i + 1;

    END LOOP;

    DBMS_OUTPUT.PUT_LINE('Final value: ' || v_i);


END;
```

## Manejo de excepciones


Con el uso de EXCEPTION podemos manejar los diferentes errores que aparecen al momento de ejecutar

* En este link podemos encontrar excepciones ya predefinidas https://www.ibm.com/docs/es/db2/11.5.x?topic=plsql-exception-handling

* Las excepciones tienen que estar manejadas desde la más especifica a la más general


```
DECLARE 
    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%rowtype;
BEGIN

    select * into v_product from PRODUCTO WHERE CODIGO = v_code;
    DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
    DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
    DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
    DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);
EXCEPTION 
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || v_code || ' was not found');

    WHEN others THEN 
        DBMS_OUTPUT.PUT_LINE('Error unidentified');

END;
```

## Manejo de excepciones propias


También se da el caso en que queremos crear nuestras propias excepciones y lanzarlas, para darle mas claridad a ciertas cosas

```
DECLARE 
    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%rowtype;
    price_limit EXCEPTION; -- le declaramos su tipo, en este casos era una excepcion
BEGIN

    select * into v_product from PRODUCTO WHERE CODIGO = v_code;

    IF v_product.precio >= 100 THEN
        RAISE price_limit;

    END IF;

    DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
    DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
    DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
    DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);

EXCEPTION 
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || v_code || ' was not found');
    
    WHEN price_limit THEN
        DBMS_OUTPUT.PUT_LINE('Limit price was reached');

    WHEN others THEN 
        DBMS_OUTPUT.PUT_LINE('Error unidentified');

END;
```

Tambien podemos lanzar una excepcion a nivel aplicacion con un mensaje de error mejor definido

```
DECLARE 
    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%rowtype;
    price_limit EXCEPTION; -- le declaramos su tipo, en este casos era una excepcion
BEGIN

    select * into v_product from PRODUCTO WHERE CODIGO = v_code;

    IF v_product.precio >= 100 THEN
        RAISE_APPLICATION_ERROR(-20998, 'Price limit of 100$ was reached'); -- el primer parametro es como el codigo de error que aparece en la consola (de -20000 hasta -20999), y el segundo parametro el mensaje de error
        -- si manejamos la exepcion en el bloque EXCEPTION (por ejemplo con un other), esta se sobrepone al RAISE_APPLICATION_ERROR
    END IF;

    DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
    DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
    DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
    DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);

EXCEPTION 
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || v_code || ' was not found');

END;
```

Tambien podemos agregarle una excepcion a un codigo de error

```
DECLARE 
    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%rowtype;
    price_limit EXCEPTION; -- le declaramos su tipo, en este casos era una excepcion
    PRAGMA EXCEPTION_INIT(price_limit, -20998);
BEGIN

    select * into v_product from PRODUCTO WHERE CODIGO = v_code;

    IF v_product.precio >= 100 THEN
        RAISE_APPLICATION_ERROR(-20998, 'Price limit of 100$ was reached'); -- el primer parametro es como el codigo de error que aparece en la consola (de -20000 hasta -20999), y el segundo parametro el mensaje de error
        -- si manejamos la exepcion en el bloque EXCEPTION (por ejemplo con un other), esta se sobrepone al RAISE_APPLICATION_ERROR
    END IF;

    DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
    DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
    DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
    DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);

EXCEPTION 
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || v_code || ' was not found');
    
    WHEN price_limit THEN
        DBMS_OUTPUT.PUT_LINE(sqlcode); -- aqui imprimimos el codigo de error

        DBMS_OUTPUT.PUT_LINE(sqlerrm); -- aquí imprimimos el error completo (codigo: mensaje)


END;
```

## Procedimientos y Funciones


La diferencia clave entre un procedimiento y una función es que las funciones nos **retornan algo**, los procedimientos por su parte no retornan nada

```
CREATE OR REPLACE PROCEDURE infoProducto(p_code producto.codigo%TYPE) 
AS 
    v_product PRODUCTO%rowtype;

BEGIN
    select * into v_product from PRODUCTO WHERE CODIGO = p_code;

    DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
    DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
    DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
    DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);   

EXCEPTION 
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || p_code || ' was not found');

    WHEN others THEN
        DBMS_OUTPUT.PUT_LINE('Error unidentified');

END;
```

Luego de compilar el procedimiento ya podemos hacer uso de él

```
DECLARE 

    v_code NUMBER(2) := &codeNumber;

BEGIN

    infoProducto(v_code);

END;
```

Ahora miremos como podemos crear funciones en PL/SQL


```
CREATE OR REPLACE FUNCTION getProduct(p_code PRODUCTO.codigo%TYPE) 
return PRODUCTO%ROWTYPE
AS
    v_product PRODUCTO%ROWTYPE;
BEGIN
    SELECT * INTO v_product FROM PRODUCTO WHERE codigo = p_code;
    
    return v_product;

EXCEPTION
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || p_code || ' was not found');
        return null;

    WHEN others THEN
        DBMS_OUTPUT.PUT_LINE('Error unidentified');
        return null;



END;

```

```
DECLARE 

    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%ROWTYPE;

BEGIN

    v_product := getProduct(v_code); -- obtenemos el producto usando la funcion
    -- Ahora hagamos una comprobación de que el dato no esté vacío
    IF v_product.codigo IS NOT NULL THEN
        DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
        DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
        DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
        DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);
    END IF;


END;
```

## Parametros de entrada y salida

En PL/SQL podemos dar un parametro como entrada "rellenarlo" de data y sin retornarlo poderlo usar en otra instancia, para esto necesitamos decir que dicho parametro será de salida, y le damos el identifier OUT



```
CREATE OR REPLACE PROCEDURE refillProduct(p_code PRODUCTO.CODIGO%TYPE, p_product OUT PRODUCTO%ROWTYPE) 
AS

BEGIN

    SELECT * INTO p_product FROM PRODUCTO WHERE codigo = p_code;
    
EXCEPTION
    WHEN no_data_found THEN
        DBMS_OUTPUT.PUT_LINE('Product with code ' || p_code || ' was not found');

    WHEN others THEN
        DBMS_OUTPUT.PUT_LINE('Error unidentified');

END;
```

Ahora veamos como sin retornarlo, podemos hacer uso de los datos agregados dentro del procedimiento

```
DECLARE 

    v_code NUMBER(2) := &codeNumber;
    v_product PRODUCTO%ROWTYPE;

BEGIN

    refillProduct(v_code, v_product); 

    -- Ahora hagamos una comprobación de que el dato no esté vacío
    IF v_product.codigo IS NOT NULL THEN
        DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
        DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
        DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
        DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);
    END IF;


END;
```

## Packages

Los paquetes nos sirven para agrupar diferentes procedimientos y funciones. Esto ayuda a tener un código más estructurado, mejor organizado y legible.


```
CREATE OR REPLACE PACKAGE Products 
AS 
    PROCEDURE dataProducts(p_code IN PRODUCTO.CODIGO%TYPE);
    FUNCTION getProducts(p_code IN PRODUCTO.CODIGO%TYPE) RETURN PRODUCTO%ROWTYPE;
END;
/
CREATE OR REPLACE PACKAGE BODY Products 
AS
    PROCEDURE dataProducts(p_code IN producto.codigo%TYPE) 
    AS 
        v_product PRODUCTO%rowtype;

    BEGIN
        select * into v_product from PRODUCTO WHERE CODIGO = p_code;

        DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
        DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
        DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
        DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);   

    EXCEPTION 
        WHEN no_data_found THEN
            DBMS_OUTPUT.PUT_LINE('Product with code ' || p_code || ' was not found');

        WHEN others THEN
            DBMS_OUTPUT.PUT_LINE('Error unidentified');

    END;

    FUNCTION getProducts(p_code IN PRODUCTO.codigo%TYPE) 
    return PRODUCTO%ROWTYPE
    AS
        v_product PRODUCTO%ROWTYPE;
    BEGIN
        SELECT * INTO v_product FROM PRODUCTO WHERE codigo = p_code;
        
        return v_product;

    EXCEPTION
        WHEN no_data_found THEN
            DBMS_OUTPUT.PUT_LINE('Product with code ' || p_code || ' was not found');
            return null;

        WHEN others THEN
            DBMS_OUTPUT.PUT_LINE('Error unidentified');
            return null;

    END;

END;
```

## Cursores

Los cursores nos sirven para almacenar más de un registro o fila de una consulta SQL. Luego podremos recorrer el cursor con un bucle para acceder a los datos. 

```
DECLARE

    CURSOR c_products IS SELECT * FROM PRODUCTO ORDER BY PRECIO;
    v_product PRODUCTO%ROWTYPE;

BEGIN

    OPEN c_products; -- Para poder utilzar el cursor tenemos que abrirlo (se puede abrir de manera implicita)

    LOOP
        FETCH c_products INTO v_product;
        EXIT WHEN c_products%notfound; -- aquí hacemos que el loop se cierre cuando no hayan más datos para leer del cursor

        DBMS_OUTPUT.PUT_LINE('  PRODUCT INFORMATION ');
        DBMS_OUTPUT.PUT_LINE('Product name: ' || v_product.nombre);
        DBMS_OUTPUT.PUT_LINE('Product code: ' || v_product.codigo);
        DBMS_OUTPUT.PUT_LINE('Product price: ' || v_product.precio);
        DBMS_OUTPUT.PUT_LINE('Product manufacturer_code: ' || v_product.codigo_fabricante);
        DBMS_OUTPUT.PUT_LINE(' ');

    END LOOP;

    CLOSE c_products; -- Cerramos el cursor para que no haya problemas

END;
```

También podemos relacionar tablas de la siguiente manera

```
DECLARE

    CURSOR c_products_manufacturer IS 
    SELECT 
        p.CODIGO,
        p.NOMBRE,
        p.PRECIO,
        f.nombre AS name_f
    FROM PRODUCTO p, FABRICANTE f 
    WHERE p.codigo_fabricante = f.CODIGO
    ORDER BY PRECIO;

    v_code PRODUCTO.CODIGO%TYPE;
    v_name PRODUCTO.NOMBRE%TYPE;
    v_price PRODUCTO.PRECIO%TYPE;
    v_namef FABRICANTE.NOMBRE%TYPE;


    
    v_products_manufacturer c_products_manufacturer%ROWTYPE;-- Esta es una forma más sencilla para este tipo de consultas, usar el cursor como base para el rowtype



BEGIN

    

    FOR v_products_manufacturer IN c_products_manufacturer LOOP

        DBMS_OUTPUT.PUT_LINE('  PRODUCT INFORMATION ');
        DBMS_OUTPUT.PUT_LINE('Product name: ' || v_products_manufacturer.nombre);
        DBMS_OUTPUT.PUT_LINE('Product code: ' || v_products_manufacturer.codigo);
        DBMS_OUTPUT.PUT_LINE('Product price: ' || v_products_manufacturer.precio);
        DBMS_OUTPUT.PUT_LINE('Product manufacturer name: ' || v_products_manufacturer.name_f);
        DBMS_OUTPUT.PUT_LINE(' ');

    END LOOP;


END;
```

## TRIGGERS

Un trigger es un bloque de codigo PL que se ejecuta justo antes o después de una operacion que especifiquemos en una tabla

Veamos el ejemplo para que se ejecute antes de insertar, borrar o actualizar una de nuestras tablas


```
CREATE OR REPLACE TRIGGER status_operation_before
BEFORE INSERT OR UPDATE OR DELETE ON PRODUCTO
BEGIN

    IF INSERTING THEN
        DBMS_OUTPUT.PUT_LINE('INSERTING DATA.....');

    ELSIF UPDATING THEN
        DBMS_OUTPUT.PUT_LINE('UPDATING DATA.....');

    ELSIF DELETING THEN
        DBMS_OUTPUT.PUT_LINE('DELETING DATA.....');

    END IF;

END;
```

Ahora un ejemplo con lo mismo pero despues de cada operacion 

```
CREATE OR REPLACE TRIGGER status_operation_after
AFTER INSERT OR UPDATE OR DELETE ON PRODUCTO
BEGIN

    IF INSERTING THEN
        DBMS_OUTPUT.PUT_LINE('DATA INSERTED');

    ELSIF UPDATING THEN
        DBMS_OUTPUT.PUT_LINE('DATA UPDATED');

    ELSIF DELETING THEN
        DBMS_OUTPUT.PUT_LINE('DATA DELETED');

    END IF;

END;
```

Ahora veamos como al realizar alguna de estas operaciones estos triggers se ejecutan de manera automatica

```

BEGIN
    INSERT INTO PRODUCTO VALUES (12, 'Portatil LCQ Gamer', 300, 2);
    
END;
```

Los trigger tienen muchos usos muy interesantes, entre ellos está el poder ver el estado de un dato antes de la operacion SQL. 

Veamos un ejemplo de esto

```
CREATE OR REPLACE TRIGGER product_validation 
BEFORE INSERT OR UPDATE ON PRODUCTO FOR EACH ROW -- Esto ultimo lo usamos para que el trigger se ejecute para cada fila que se insertará (si son 5 se ejecutaria 5 veces)
DECLARE

    v_num_manuf NUMBER(2);

BEGIN

    IF :new.precio < 0 THEN -- El for each row nos desbloquea :new o :old para verificar el dato actual antes de la operacion
        RAISE_APPLICATION_ERROR(-20002, ' Price under the expected');
    END IF;


    -- Comprobaremos si el numero de fabricante dado existe en la tabla fabricantes
    SELECT COUNT(*) INTO v_num_manuf FROM FABRICANTE WHERE CODIGO = :new.CODIGO_FABRICANTE;

    if v_num_manuf <= 0 THEN
        RAISE_APPLICATION_ERROR(-20003, ' Manufacuter code ' || :new.CODIGO_FABRICANTE || ' doesnt exist');
    END IF; 

END;
```

**NOTA:** Cuando se usa el FOR EACH ROW bloqueamos **TODAS** las operaciones en esa tabla en la que se está operando

Veamos como podemos usar un trigger para situaciones donde un campo va a ser actualizado de manera incorrecta

```
CREATE OR REPLACE TRIGGER update_manufactuer 
BEFORE UPDATE OF CODIGO_FABRICANTE ON PRODUCTO FOR EACH ROW
BEGIN

    if :new.CODIGO_FABRICANTE <> :old.CODIGO_FABRICANTE THEN
        :new.PRECIO := :new.PRECIO + 10;
    END IF;

END; 
```
